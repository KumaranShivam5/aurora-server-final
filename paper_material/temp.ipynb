```python
df = pd.read_csv('../../not_on_git/final_classification/all_classified.csv', index_col='name')
lx = df[df['class']=='LMXB']
hx = df[df['class']=='HMXB']
lx 
```

```python
hx.to_csv('temp_files/HMXB.csv')
```

```python
plt.hist(lx['prob'], density=True)
plt.show()
```

```python
plt.hist(hx['prob'], density=True)
plt.show()
```

```python
lx_low = lx.query("prob>0.35 and prob < 0.55")
lx_high = lx.query("prob>0.85")
```

```python
hx_low = hx.query("prob>0.35 and prob < 0.55")
hx_high = hx.query("prob>0.85")
```

```python
hx_low['b-csc'].isna().sum() / len(lx_low)*100
```

```python
import numpy as np 
```

```python
for c in lx.columns.tolist()[:2]:
    fig , ax = plt.subplots(nrows = 1 , ncols=2, figsize = (16,6))
    ax = np.ravel(ax)
    lx_low_count = int((lx_low[c].isna().sum()))
    lx_high_count = int((lx_high[c].isna().sum()))
    hx_low_count = int((hx_low[c].isna().sum()))
    hx_high_count = int((hx_high[c].isna().sum()))
    ax[0].hist(lx_low[c], histtype='step' , label=f'low : {lx_low_count} / {len(lx_low)}')
    ax[0].hist(lx_high[c] ,  histtype='step', label=f'High: {lx_high_count} / {len(lx_high)}')
    ax[1].hist(hx_low[c], histtype='step' , label=f'low : {hx_low_count} / {len(hx_low)}')
    ax[1].hist(hx_high[c] ,  histtype='step', label=f'High: {hx_high_count} / {len(hx_high)}')
    for a in ax:
        a.legend(fontsize=16)
        a.set_xlabel(c)
    ax[0].set_title('LMXB')
    ax[1].set_title('HMXB')
    # plt.savefig(f'temp_files/temp_plots/{c}_lmxb.jpg')
    plt.show()
    plt.close()
```

```python
for c in hx.columns.tolist():
    low_count = int((hx_low[c].isna().sum()) )
    high_count = int((hx_high[c].isna().sum()))
    plt.hist(hx_low[c], histtype='step' , label=f'low : {low_count} / {len(hx_low)}')
    plt.hist(hx_high[c] ,  histtype='step', label=f'High: {high_count} / {len(hx_high)}')
    plt.legend()
    plt.savefig(f'temp_files/temp-hmxb_plots/{c}_hmxb.jpg')
    plt.close()
```
