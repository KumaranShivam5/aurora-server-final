```python
df = pd.read_csv('../../not_on_git/final_classification/all_classified.csv', index_col='name')
lx = df[df['class']=='LMXB']
hx = df[df['class']=='HMXB']
lx 
```

```python
plt.hist(lx['prob'], density=True)
plt.show()
```

```python
plt.hist(hx['prob'], density=True)
plt.show()
```

```python
lx_low = lx.query("prob>0.35 and prob < 0.55")
lx_high = lx.query("prob>0.85")
```

```python
hx_low = hx.query("prob>0.35 and prob < 0.55")
hx_high = hx.query("prob>0.85")
```

```python
hx_low['b-csc'].isna().sum() / len(lx_low)*100
```

```python
low_count = int((lx_low['b-csc'].isna().sum()) )
high_count = int((lx_high['b-csc'].isna().sum()))
plt.hist(lx_low['b-csc'], histtype='step' , label=f'low : {low_count} / {len(lx_low)}')
plt.hist(lx_high['b-csc'] ,  histtype='step', label=f'High: {high_count} / {len(lx_high)}')
plt.legend()
plt.show()
```

```python
low_count = int((hx_low['var_inter_index_b'].isna().sum()) )
high_count = int((hx_high['var_inter_index_b'].isna().sum()))
plt.hist(hx_low['var_inter_index_b'], histtype='step' , label=f'low : {low_count} / {len(hx_low)}')
plt.hist(hx_high['var_inter_index_b'] ,  histtype='step', label=f'High: {high_count} / {len(hx_high)}')
plt.legend()
plt.show()
```

```python
for c in lx.columns.tolist():
    low_count = int((lx_low[c].isna().sum()) )
    high_count = int((lx_high[c].isna().sum()))
    plt.hist(lx_low[c], histtype='step' , label=f'low : {low_count} / {len(lx_low)}')
    plt.hist(lx_high[c] ,  histtype='step', label=f'High: {high_count} / {len(lx_high)}')
    plt.legend()
    plt.savefig(f'temp_files/temp_plots/{c}_lmxb.jpg')
    plt.close()
```

```python
for c in hx.columns.tolist():
    low_count = int((hx_low[c].isna().sum()) )
    high_count = int((hx_high[c].isna().sum()))
    plt.hist(hx_low[c], histtype='step' , label=f'low : {low_count} / {len(hx_low)}')
    plt.hist(hx_high[c] ,  histtype='step', label=f'High: {high_count} / {len(hx_high)}')
    plt.legend()
    plt.savefig(f'temp_files/temp-hmxb_plots/{c}_hmxb.jpg')
    plt.close()
```
