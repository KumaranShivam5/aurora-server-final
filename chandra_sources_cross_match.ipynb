```python
plt.hist(sep2d_.arcsec, histtype='step',
         bins=np.logspace(-2, 2., 64))
plt.xlabel('separation [arcsec]')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.show()
```

```python
cm = pd.DataFrame({
    'idx' : idx_ , 
    'sep' : sep2d_.arcsec
})
```

```python
df_agn_csc = df[df['class']=='AGN']
df_agn_csc = df_agn_csc.sort_values(by='offset')
display(df_agn_csc)
df_agn_csc = df_agn_csc.drop_duplicates('name')
df_agn_csc = df_agn_csc.drop_duplicates('cat_name')
df_agn_csc = df_agn_csc.drop_duplicates(['ra' , 'dec'])
df_agn_csc = df_agn_csc.drop_duplicates(['cat_ra' , 'cat_dec'])
df_agn_csc.to_csv('data_v3/csc_AGN.csv')
df_agn_csc
```
